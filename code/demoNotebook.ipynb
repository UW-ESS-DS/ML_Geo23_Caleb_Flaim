```python
# import required modules
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy as sp 
import random
```

```python
# make a function that accepts two lists as arguments and plots them against eachother

def makePlot (list1, list2):
    """This function accepts two lists or numpy arrays as arguments and plots them against eachother. 
    It then saves the plot as a PNG to the project directory.
    """

    fig, axs = plt.subplots(1,2, figsize = (11,8)) # make figure and axes objects

    # make and store plot attibutes in lists
    titles = ['This plots list2 v.s. list1 as a line plot', "This plots list1 and list2 in a histogram"]
    x_labels = ['This is the x-axis label', 'This is the x-axis label']
    y_labels = ['This is the y-axis label', 'This is the y-axis label']
    # do the plotting
    for i, ax in enumerate(axs):
        if i:
            ax.plot(list1, list2, c='salmon', label='data')
        else:
            ax.hist(list1, bins=100, color='salmon', alpha=.5, label='List1')
            ax.hist(list2, bins=100, color='olive', alpha=.5, label='List2')
        ax.set_title(titles[i])
        ax.set(xlabel=x_labels[i], ylabel=y_labels[i])
        ax.legend()

    plt.savefig("/Users/calebflaim/Documents/ML_Geo23_Caleb_Flaim/plots/examplePLot.png") # adjust filepath for your system!
    plt.show()
```

```python
# make lists

def makeRandomLists(list_length):
    """
    This takes 1 argument:
    list_length

    and returns two randoms lists of this length
    """

    list1 = []
    list2 = []

    for i in range(list_length):
        list1.append(random.random())
        list2.append(random.random())

    return list1, list2
```

```python
# make lists and then pass them to the plotting function

list1, list2 = makeRandomLists(1000)

makePlot(list1, list2)
```
